```python
# Import necessary packages
import os
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
import rioxarray as rxr 
# Plotting extent is used to plot raster & vector data together
from rasterio.plot import plotting_extent

import earthpy as et
import earthpy.plot as ep

# Prettier plotting with seaborn
sns.set(font_scale=1.5, style="white")
```

```python
# Get data and set working directory
et.data.get_data("colorado-flood")
os.chdir(os.path.join(et.io.HOME, 'earth-analytics', 'data'))
```

```python
# Load the pre_DTM_hill.tif layer
file_path = "colorado-flood/spatial/boulder-leehill-rd/pre-flood/lidar/pre_DTM_hill.tif"
file_path2 = "colorado-flood/spatial/boulder-leehill-rd/pre-flood/lidar/pre_DTM.tif"
pre_dtm_hill = rxr.open_rasterio(file_path)
dtm_pre_arr = rxr.open_rasterio(file_path2)
# Plot the data using ep.plot_bands()
ep.plot_bands(pre_dtm_hill, cmap="gray",title="Challenge 1: Lidar Digital Terrain Model (DTM) Overlayed on Top of Hillshade")
```

```python
# Overlay the DTM on the hillshade layer
fig, ax = plt.subplots(figsize=(10, 6))
ep.plot_bands(dtm_pre_arr, cmap="viridis", ax=ax,title="Challenge 2: Lidar Digital Terrain Model (DTM) Overlayed on Top of Hillshade")
ep.plot_bands(pre_dtm_hill, cmap="gray",alpha=0.5,ax=ax,cbar="False")


ax.set_axis_off()
plt.show()
```
